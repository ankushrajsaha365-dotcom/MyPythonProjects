```python
try:
    a = int(input())
    b = int(input())
    a/b
except ZeroDivisionError:
    print("cannot be divided by zero!")
else:
    print(a/b)
```

```python
try:
    x = int(input("Enter number: "))
    print(10 / x)
except ValueError:
    print("Invalid input! Enter an integer.")
except ZeroDivisionError:
    print("Division by zero not allowed.")
```

```python
try:
    f = open("data.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("File not found")
finally:
    print("Program finished")


### There's a data.txt file in the same directory , so the except block isn't applied
```

```python
marks = int(input("Enter marks: "))

if marks < 0 or marks > 100:
    raise ValueError("Marks must be between 0 and 100")

print("Marks accepted")
```

```python
class AgeError(Exception):
    pass
try:
    age = int(input("Enter your age:"))
    if age not in range(1,100):
        raise AgeError("Age not defined well!")
    print("Age is accepted")

except AgeError as ae:
    print(ae)

finally:
    print("Program completed")
```
