# Tuples

```python
# Declare a tuple
my_tuple = (1, 2, 3, 'hello', 4.5)
print(my_tuple)
```

### Tuple Methods

```python
tup = (1 , 2 , 4 , 'Hi' , 'Python' , 2)
print(tup.count(2))                 # Number of occurrences
print(tup.index(2))                 # Index of first occurrence
```

__Note: Methods like append(), remove(), or sort()__

__do not exist for tuples because they are immutable.__


### Built-in Functions with Tuples


```python
tup = (1,2,3,4,5,6)

print(len(tup))
print(max(tup))
print(min(tup))
print(sum(tup))
```

```python
my_tuple = (11,54,12,76,3,98,34)

print(sorted(my_tuple))                # sorted(tuple) → Returns a new sorted list
```

```python
my_tuple = (True , False , False)
print(any(my_tuple))                # any(tuple) → True if any element is True
print(all(my_tuple))                # all(tuple) → True if all elements are True
```

```python
nums = (11,23,34,45,56,67,78)

for index, value in enumerate(nums):
    print(index, value)

```

```python
t1 = (12,32,34,56,86)
t2 = (1,9,3,6,7)

print(zip(t1,t2))               # zip(tuple1, tuple2) → Combine multiple iterables

# It returns a zip object, which is an iterator.
# When you print it directly, Python just shows its memory location

print(tuple(zip(t1,t2)))        # tuple(iterable) → Converts iterable to tuple
```
