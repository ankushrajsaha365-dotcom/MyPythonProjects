# Dictionary

### Dictionary Methods

```python
d = {"name": "abc" ,"age": 20}
d.keys()
```

```python
print(d.values())
```

```python
print(d.items())
```

```python
print(d.get("name"))
```

```python
print(d.get("class"))
```

```python
d.update({"grade":"A"})
print(d)
```

```python
d1 = d.copy()

d1.pop("age")            # **pop(key, default)** → Removes key and returns its value
```

```python
d2 = d.copy()

d2.popitem()            # **popitem()** → Removes and returns an arbitrary (key, value) pair
```

### Built-in functions

```python
my_dict = {"name": "abc", "age": 20 ,"grade": "A"}
print(len(my_dict))
print(sorted(my_dict))
```

```python
pairs = [("a", 1), ("b", 2), ("c", 3)]
d = dict(pairs)
# {'a': 1, 'b': 2, 'c': 3}
print(d)
```

```python
keys = ["a", "b", "c"]
values = [1, 2, 3]
dict(zip(keys, values))
```
