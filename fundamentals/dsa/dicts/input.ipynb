```python
d = {}

n = int(input("How many items? "))

for _ in range(n):
    key = input("Key: ")
    value = input("Value: ")
    d[key] = value

print(d)
```

```python
data = input("Enter key value pairs: ").split()

d = dict(zip(data[::2], data[1::2]))
print(d)



# take input like a 1 b 2 c 3
```
