üîπ 3Ô∏è‚É£ __range()__

Generates a sequence of numbers.

__Forms__

- range(stop)
- range(start, stop)
- range(start, stop, step)


```python
print(list(range(5)))
# [0, 1, 2, 3, 4]

print(list(range(1, 10, 2)))
# [1, 3, 5, 7, 9]
```

```python
nums = [5, 2, 9, 1]

print(min(nums))  # 1
print(max(nums))  # 9
print(min("raj"))  # 'a'   # works on strings too

```

üîπ 8Ô∏è‚É£ __sorted()__

Returns new sorted list.

Important:
- sorted() does NOT modify original list.

```python
nums = [4, 1, 3, 2]
print(sorted(nums))             # By default ascending

print(sorted(nums, reverse=True))   # descending
```

üîπ 9Ô∏è‚É£ __enumerate()__

Gives index + value.

```python
names = ["Raj", "Aman", "Riya"]

for index, name in enumerate(names):
    print(index, name)


for index, name in enumerate(names, start = 1):         # custom start
    print(index, name)
```
