```python
import numpy as np

# Indexing and slicing

arr = np.array([10, 20, 30, 40, 50])

print(arr[0])
print(arr[-1])
print(arr[1:4])
```

```python
# 2D 
arr = np.array([[1,2,3],[4,5,6]])

print(arr[0, 1])   # row 0, column 1

print(arr)
```

```python
arr = np.array([1, 2, 3, 4])
print(arr ** 2)
```

```python
# Even or odd

arr = np.array([1, 2, 3, 4, 5, 6])
print(arr[arr % 2 == 0])
```
