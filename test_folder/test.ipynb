```python
class Student:
    def __init__(self,name,roll):
        self.roll = roll
        self.name = name
    def display(self):
        print(f"Student id {self.roll}")

name = input("Enter name:")
roll = int(input("Enter student id:"))
student = Student(name,roll)
student.display()
```
